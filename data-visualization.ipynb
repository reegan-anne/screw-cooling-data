```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```
data_dir = '/content/drive/My Drive/Screw Cooling in an Annular Channel/Data_Friday/50 with blower on 5ms'
merged_data_5ms = pd.DataFrame()

for filename in os.listdir(data_dir):
    print(filename)
    dataset = pd.read_csv(os.path.join(data_dir, filename), 
                sep='\t', skiprows=4 , encoding='UTF-16 LE')
    dataset_mean = np.array(dataset.mean())
    dataset_mean = pd.DataFrame(dataset_mean.reshape(1,22))
    merged_data_5ms = merged_data_5ms.append(dataset_mean)

merged_data_5ms.columns = ["Time", "101 (C)", "102 (C)", "103 (C)", "104 (C)", "105 (C)",
                   "106 (C)", "107 (C)", "108 (C)", "109 (C)", "110 (C)", "111 (C)", "112 (C)",
                   "113 (C)", "114 (C)", "115 (C)", "116 (C)", "117 (C)", "118 (VDC)", "201 <velocity> (VDC)",	"203 <test section> (VDC)",	"204 <pipe exit> (VDC)"]
merged_data_5ms.loc['mean'] = merged_data_5ms.mean()
```

```
df_thermoc = merged_data_5ms.drop([ "Time", "103 (C)","117 (C)", "118 (VDC)", "201 <velocity> (VDC)",	"203 <test section> (VDC)",	"204 <pipe exit> (VDC)"], axis = 1)
df_thermoc_av = df_thermoc.mean(axis=0)
```

```
df_thermoc_av = df_thermoc_av.reindex(index=df_thermoc_av.index[::-1])
```

```
df_thermoc_av = pd.DataFrame(df_thermoc_av)
```

```
df_thermoc_av
```

```
df_thermoc_av.index
```

```
df_thermoc_av.plot(figsize=(10,6))
plt.xlabel("Thermocouple");
plt.ylabel("Temperature (C)");
plt.title("Temperature drop 5m/s");
plt.grid();
plt.xticks(np.arange(15),df_thermoc_av.index);
```
