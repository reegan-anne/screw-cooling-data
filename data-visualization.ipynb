```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/drive/My Drive/Screw Cooling in an Annular Channel/Data_Friday/50 with blower on/10ms test 1.csv', sep='\t', skiprows=4 , encoding='UTF-16 LE')
df.columns = ["Scan", "Time", "101 (C)", "102 (C)", "103 (C)", "104 (C)", "105 (C)", "106 (C)", "107 (C)", "108 (C)", "109 (C)", "110 (C)", "111 (C)", "112 (C)", "113 (C)", "114 (C)", "115 (C)", "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"]
df.set_index("Scan")
```

```
df_thermoc = df.drop([ "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"], axis = 1)
df_thermoc_av = df_thermoc.mean(axis=0)
df_thermoc_av = df_thermoc_av.drop("Scan")
```

```
df_thermoc_av = df_thermoc_av.reindex(index=df_thermoc_av.index[::-1])
```

```
df_thermoc_av.plot()
plt.xlabel("Thermocouple")
plt.ylabel("Temperature (C)")
plt.title("Temperature drop 5m/s Test 1")
```

# Averaging all three tests

```
df2 = pd.read_csv('/content/drive/My Drive/Screw Cooling in an Annular Channel/Data_Friday/50 with blower on/10ms test 2.csv', sep='\t', skiprows=4 , encoding='UTF-16 LE')
df2.columns = ["Scan", "Time", "101 (C)", "102 (C)", "103 (C)", "104 (C)", "105 (C)", "106 (C)", "107 (C)", "108 (C)", "109 (C)", "110 (C)", "111 (C)", "112 (C)", "113 (C)", "114 (C)", "115 (C)", "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"]
df2.set_index("Scan")
```

```
df2_thermoc = df2.drop([ "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"], axis = 1)
df2_thermoc_av = df2_thermoc.mean(axis=0)
df2_thermoc_av = df2_thermoc_av.drop("Scan")
df2_thermoc_av = df2_thermoc_av.reindex(index=df2_thermoc_av.index[::-1])
```

```
df3 = pd.read_csv('/content/drive/My Drive/Screw Cooling in an Annular Channel/Data_Friday/50 with blower on/10ms test 3.csv', sep='\t', skiprows=4 , encoding='UTF-16 LE')
df3.columns = ["Scan", "Time", "101 (C)", "102 (C)", "103 (C)", "104 (C)", "105 (C)", "106 (C)", "107 (C)", "108 (C)", "109 (C)", "110 (C)", "111 (C)", "112 (C)", "113 (C)", "114 (C)", "115 (C)", "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"]
df3.set_index("Scan")
```

```
df3_thermoc = df3.drop([ "116 (C)", "117 (C)", "118 (VDC)", "119 (VDC)", "120 (VDC)"], axis = 1)
df3_thermoc_av = df3_thermoc.mean(axis=0)
df3_thermoc_av = df3_thermoc_av.drop("Scan")
df3_thermoc_av = df3_thermoc_av.reindex(index=df3_thermoc_av.index[::-1])
```

# Joining all three tests

```
df_total = pd.concat([df_thermoc_av, df2_thermoc_av, df3_thermoc_av], axis=1, ignore_index=True)
```

```
df_total["10m/s"] = df_total.mean(axis=1)
```

```
df_total
```

```
index = df_total.index
df_total.plot(y="10m/s", kind="line", figsize = (14,6), use_index = True)
plt.xlabel("Thermocouple")
plt.xticks(list(range(len(index))), df_total.index)
plt.ylabel("Temperature (C)")
plt.title("Temperature drop 10m/s with Blower on")
```
